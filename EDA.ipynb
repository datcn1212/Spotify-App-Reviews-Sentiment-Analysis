## Spotify App Reviews Sentiment Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
stf = pd.read_csv('reviews.csv')
```

```python
ax = stf['Rating'].plot(kind="hist", color="slategrey", bins=30, figsize=(4, 4), fontsize=8, rot=0)
ax.set_title("Number of reviews by Rating", fontsize=10)
ax.set_xlabel("rating", fontsize=8)
ax.set_ylabel("number", fontsize=8)
```

```python
cl = sns.color_palette('pastel')[0:5]
plt.pie(stf['Rating'].value_counts(), labels = stf['Rating'].unique(), colors = cl, autopct='%1.1f%%')
plt.show()
```

```python
stf1 = stf
rate = [1,2,3,4,5]
sentiment = ['negative', 'negative', 'positive', 'positive', 'positive']
for i in range(5):
    stf1["Rating"].replace(rate[i], sentiment[i],inplace=True)
cl = sns.color_palette('pastel')[0:5]
plt.pie(stf1['Rating'].value_counts(), labels = stf1['Rating'].unique(), colors = cl, autopct='%1.1f%%')
plt.show()
```

```python
stf['len'] = stf['Review'].str.len()
sns.histplot(stf['len'][stf.len<600])
plt.title('Lenght of Reviews')
plt.xlabel('')
plt.show()
```

```python
review = stf['Review']
plt.figure(figsize=(5,5))
wc = WordCloud(max_words=5000, min_font_size=6, height=1000, width=2000, collocations=False,background_color='#C5E0F6').generate(str(review))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
